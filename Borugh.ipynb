```python
import pandas as pd

# Step 1: Load the data from the Excel file
previous_week_df = pd.read_excel(r"C:\Users\User\Downloads\Cash_Position_Mar10_2025 (1).xlsx", sheet_name="Aging Detail", header=6)

# Step 2: Drop columns that start with 'Unnamed' (which usually are empty or irrelevant columns)
previous_week_df = previous_week_df.loc[:, ~previous_week_df.columns.str.startswith('Unnamed')]

# Step 3: Identify rows with Shop Name (non-numeric values in 'Invoice Number')
previous_week_df['Is_Shop'] = ~previous_week_df['Invoice Number'].astype(str).str.isnumeric()

# Step 4: Forward fill 'Shop Name' based on 'Invoice Number' rows
previous_week_df['Shop Name'] = previous_week_df['Invoice Number'].where(previous_week_df['Is_Shop']).ffill()

# Step 5: Keep only rows with actual invoice numbers (non-Shop Name rows)
previous_week_df = previous_week_df[~previous_week_df['Is_Shop']].copy()

# Optional: Drop the temporary 'Is_Shop' column after filtering
previous_week_df.drop(columns='Is_Shop', inplace=True)

# Step 6: Drop rows where 'Invoice Date' is missing (NaN)
previous_week_df = previous_week_df[previous_week_df['Invoice Date'].notna()]

# Step 7: Group by 'Shop Name' and sum the specified columns
previous_week_df_grouped = previous_week_df.groupby('Shop Name')[['Balance', '0-30 Days', '31-60 Days', '61-90 Days', 'Over 90 Days']].sum().reset_index()

# Step 8: Print the result
result = previous_week_df_grouped['Balance'].sum()
print(result)
```

```python
import pandas as pd

# Step 1: Load the data from the Excel file
current_week_df = pd.read_excel(r"C:\Users\User\Downloads\Cash_Position_Mar17_2025 (1).xlsx", 
                                sheet_name="Aging Detail", header=6)

# Step 2: Drop columns that start with 'Unnamed' (which usually are empty or irrelevant columns)
current_week_df = current_week_df.loc[:, ~current_week_df.columns.str.startswith('Unnamed')]

# Step 3: Identify rows with Shop Name (non-numeric values in 'Invoice Number')
current_week_df['Is_Shop'] = ~current_week_df['Invoice Number'].astype(str).str.isnumeric()

# Step 4: Forward fill 'Shop Name' based on 'Invoice Number' rows
current_week_df['Shop Name'] = current_week_df['Invoice Number'].where(current_week_df['Is_Shop']).ffill()

# Step 5: Keep only rows with actual invoice numbers (non-Shop Name rows)
current_week_df = current_week_df[~current_week_df['Is_Shop']].copy()

# Optional: Drop the temporary 'Is_Shop' column after filtering
current_week_df.drop(columns='Is_Shop', inplace=True)

# Step 6: Drop rows where 'Invoice Date' is missing (NaN)
current_week_df = current_week_df[current_week_df['Invoice Date'].notna()]

# Step 7: Group by 'Shop Name' and sum the specified columns
current_week_df_grouped = current_week_df.groupby('Shop Name')[['Balance', '0-30 Days', '31-60 Days', '61-90 Days', 'Over 90 Days']].sum().reset_index()

# Step 8: Print the result
result = current_week_df_grouped['Balance'].sum()
print(result)
```

```python
# New entry Set 

# Assuming you have both current_week_df and previous_week_df already loaded

# Step 1: Create a set of invoice numbers from previous week
previous_week_invoice_numbers = set(previous_week_df['Invoice Number'])

# Step 2: Add a 'New Entry' column to current_week_df, where new invoice numbers will be labeled as 'New Entry'
current_week_df['New Entry'] = current_week_df['Invoice Number'].apply(
    lambda x: 'New Entry' if x not in previous_week_invoice_numbers else ''
)

# Filter the rows where 'New Entry' column has 'New Entry' label
new_entries_df = current_week_df[current_week_df['New Entry'] == 'New Entry']

# Print only the new entry rows
print(new_entries_df)

```

```python
# Step 1: Create a set of invoice numbers from the current week
current_week_invoice_numbers = set(current_week_df['Invoice Number'])

# Step 2: Add a 'Paid Entry' column to previous_week_df, where invoices not found in the current week are labeled as 'Paid Entry'
previous_week_df['Paid Entry'] = previous_week_df['Invoice Number'].apply(
    lambda x: 'Paid Entry' if x not in current_week_invoice_numbers else ''
)

# Step 3: Filter the rows where 'Paid Entry' column has 'Paid Entry' label
paid_entries_df = previous_week_df[previous_week_df['Paid Entry'] == 'Paid Entry']

# Step 4: Print only the paid entry rows
print(paid_entries_df)
```

```python
# Ensure both DataFrames have the same Shop Name order and index
merged_df = pd.merge(current_week_df_grouped, previous_week_df_grouped, on='Shop Name', how='outer', suffixes=('_current', '_previous'))

# Subtract the corresponding columns to find the difference
columns_to_diff = ['Balance', '0-30 Days', '31-60 Days', '61-90 Days', 'Over 90 Days']
for col in columns_to_diff:
    merged_df[f'{col}_diff'] = merged_df[f'{col}_current'] - merged_df[f'{col}_previous']

# Print the resulting DataFrame with the differences
print(merged_df[['Shop Name'] + [f'{col}_diff' for col in columns_to_diff]])
```

```python
import pandas as pd

# Define file path
file_path = r'C:\Users\User\Downloads\output_file.csv'

# Write new entries to the CSV file (starting at the beginning)
new_entries_df.to_csv(file_path, index=False, header=True)

# Add 2 blank lines after the new entries section
with open(file_path, 'a') as f:
    f.write('\n\n')  # Writing 2 blank lines

# Append paid entries with 2 blank lines before it
paid_entries_df.to_csv(file_path, index=False, header=True, mode='a')

# Add 2 blank lines after the paid entries section
with open(file_path, 'a') as f:
    f.write('\n\n')  # Writing 2 blank lines

# Append the merged DataFrame (differences) with 2 blank lines before it
merged_df.to_csv(file_path, index=False, header=True, mode='a')

# Now, the output CSV will have each section with 2 blank lines between them
```
